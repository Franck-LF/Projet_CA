## Imports

```python
# from PyPDF2 import PdfFileReader 
from PIL import Image, ImageDraw 
from reportlab.lib.pagesizes import letter 
from reportlab.pdfgen import canvas 
import tempfile
import fitz  # PyMuPDF
```

```python
input_file = "PDF/CARIO.pdf"
output_path = "PDF/"
output_file = "PDF/test.png"
```

## First : One page

```python
# A single page
doc = fitz.open(input_file)
page = doc[0]
pix = page.get_pixmap()
pix.save(output_file)

print("Conversion terminée : page1.png")
```

```python
input_file = "PDF/202.pdf"
doc = fitz.open(input_file)

for i, page in enumerate(doc, start=1):
    pix = page.get_pixmap()
    output = output_path + f"page_{i}.png"
    pix.save(output)
    print(f"Page {i} enregistrée sous {output}")

print("✅ Conversion terminée !")
```

```python
doc = fitz.open("PDF/202.pdf")

# Liste pour stocker les images des pages
images = []

# Zoom pour la résolution (1 = normal, 2 = double, etc.)
zoom = 1.5
matrix = fitz.Matrix(zoom, zoom)

# Conversion de chaque page en image PIL
for page in doc:
    pix = page.get_pixmap(matrix=matrix)
    img = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)
    images.append(img)

# Calcul de la hauteur totale et largeur max
total_height = sum(img.height for img in images)
max_width = max(img.width for img in images)

# Créer une grande image blanche
final_img = Image.new("RGB", (max_width, total_height), "white")

# Coller chaque page l’une sous l’autre
y_offset = 0
for img in images:
    final_img.paste(img, (0, y_offset))
    y_offset += img.height

# Sauvegarder dans un seul fichier PNG
final_img.save(output_path + "pdf_complet.png")

print("✅ Toutes les pages ont été combinées dans pdf_complet.png")
```

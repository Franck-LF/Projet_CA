## Imports

```python
import os
import re
import time

import numpy as np
import pandas as pd

from spacy.tokenizer import Tokenizer
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer
from langchain.text_splitter import RecursiveCharacterTextSplitter

import faiss

#pd.set_option('display.max_columns', 5)
pd.set_option('display.max_colwidth', 100)
pd.set_option('display.max_rows', 20)

model = SentenceTransformer(r"C:\Users\S589669\Documents\DEV\ABP-Doctheque\sentence-camembert-large", local_files_only=True)
```

## **Loading, chunking and embedding the data**

```python
df = pd.read_csv("CSV/documents_cleaned.csv", index_col = [0])

titles = df["Titles_Cleaned"].values
documents = df["Documents_Cleaned"].values

# 1. Configuration
class Config:
    # MODEL_NAME = 'all-MiniLM-L6-v2'
    CHUNK_SIZE = 256                 # Taille des chunks en tokens
    CHUNK_OVERLAP = 48               # Chevauchement entre chunks
    INDEX_FILE = "document_index.faiss"
    METADATA_FILE = "metadata.json"
    DOCUMENTS_DIR = "cleaned_docs"   # Dossier contenant les documents

# 3. Découpeur de texte
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=Config.CHUNK_SIZE,
    chunk_overlap=Config.CHUNK_OVERLAP,
    length_function=len
)

# @np.vectorize
def light_cleaning(document):
    ''' light cleaning for embedding '''
    document = document.replace('«', ' ').replace('»', ' ').replace('>', ' ').replace('<', ' ')
    document = document.replace('-', ' ').replace('+', ' ').replace('=', ' ').replace("n°", ' ')
    document = document.replace(':', ' ').replace(',', ' ').replace('.', ' ')
    document = document.replace('€', ' ')
    # Remove non-breaking spaces
    document = document.replace("\xa0", ' ')
    document = re.sub(r" +", ' ', document).strip()
    return document

def get_chunks_and_mapping_from_documents(documents, verbose = 0):
    ''' chunk a list of documents
        Return:
         - chunks: list of chunks
         - mapping: list with index of document for each chunk
    '''
    chunks = []
    mapping = []
    for i, document in enumerate(documents):
        chunks_temp = text_splitter.split_text(document)
        if verbose == 1:
            print("Nb chunks:", len(chunks_temp))
        chunks.extend(chunks_temp)
        mapping.extend([i] * len(chunks_temp))
    return chunks, mapping

# Cleaning the documents
cleaned_documents = list(map(light_cleaning, documents))
# Chunking the documents
chunks, mapping_chunk_to_document = get_chunks_and_mapping_from_documents(cleaned_documents, verbose = 0)
# chunks embedding
chunks_embedding = model.encode(chunks, convert_to_numpy=True)
```

## Using FAISS to store the embedding and search through

We try FAISS with the common indexation and search tools:
- **Flat**: 
     1. **IP** (inner product)
     2. **L2** (Euclidean distance)
- **IVF** (INverted File)<br>
       (data clustering using Kmeans and then search through central points)<br>
     1. **IVF Flat IP and L2**
     2. **IVFPQ** (add quantization but *probably useless here because too few data*)
- **HNSW** (Hierarchical Navigable Small Worlds)<br>
         *Method probably useless here because too few data*

```python
class Search_Faiss():

    def __init__(self, titles, chunks, mapping_chunk_to_document, model, embedding, index_struct = 'FlatIP'):
        ''' Initialization '''
        self.titles = titles
        self.chunks = chunks
        self.mapping_chunk_to_document = mapping_chunk_to_document
        self.model = model
        self.embedding = embedding
        self.normalize = True
        if self.normalize:
            faiss.normalize_L2(self.embedding)
        self.embedding_dimension = embedding[0].shape[0]
        self.n_cells = len(documents)
        self.n_probe = 5
        self.index = None
        self.create_index(index_struct)
        assert self.index

    def reset_index(self):
        self.index.reset()
        self.index = None

    def create_index(self, index_struct = 'FlatIP'):
        ''' Create index FAISS object '''
        if index_struct == 'FlatIP':
            self.index = faiss.IndexFlatIP(self.embedding_dimension)
        elif index_struct == 'FlatL2':
            self.index = faiss.IndexFlatL2(self.embedding_dimension)
        elif index_struct == 'IVFFlatIP':
            quantizer = faiss.IndexFlatIP(self.embedding_dimension)
            self.index = faiss.IndexIVFFlat(quantizer, self.embedding_dimension, self.n_cells)
            self.index.nprobe = self.n_probe
            self.index.train(self.embedding)
        elif index_struct == 'IVFFlatL2':
            quantizer = faiss.IndexFlatL2(self.embedding_dimension)
            self.index = faiss.IndexIVFFlat(quantizer, self.embedding_dimension, self.n_cells, faiss.METRIC_L2)
            self.index.nprobe = self.n_probe
            self.index.train(self.embedding)
        elif index_struct == 'HNSWFlat':
            M = 32 # nombre de connexions par vecteur   
            self.index = faiss.IndexHNSWFlat(self.embedding_dimension, M)
            self.index.train(self.embedding)
        
        self.index.add(self.embedding)

    def search_from_Faiss(self, queries, top_k = 5, verbose = 0):
        queries_embedding = self.model.encode(queries)
        if self.normalize:
            faiss.normalize_L2(queries_embedding)
        D, I = self.index.search(queries_embedding, k = len(self.chunks))
        for i in range(len(queries)):
            index_docs = []
            for index_chunk in I[i]:
                index_doc = self.mapping_chunk_to_document[index_chunk]
                if index_doc in index_docs:
                    continue
                index_docs.append(index_doc)
                if len(index_docs) == top_k:
                    break
            if verbose == 1 and len(index_docs):
                print(queries[i])
                for index_doc in index_docs:
                    print(f" - {self.titles[index_doc]}")

queries = [
    "dérogations",
    "animateur analyste.",
    "dérogations hors norme.",
    "règlement des impayés",
    "impayés",
    "impayé",
    "multi détention",
    "gestion des doublons de dossier",
    "résiliations",
]

for struc in [
    'FlatIP',
    'FlatL2',
    'IVFFlatIP',
    'IVFFlatL2',
    'HNSWFlat',
    ]:
    search_Faiss = Search_Faiss(titles, chunks, mapping_chunk_to_document, model, chunks_embedding, index_struct=struc)
    start_time = time.time()
    search_Faiss.search_from_Faiss(queries, top_k = 5, verbose = 1)
    end_time = time.time() - start_time
    print(f"Time: {1000 * end_time:.2f} ms")
    print("\t\t----------------")
    search_Faiss.reset_index()
```

## Annexe

1. **faiss.IndexIVFFlat**: This is similar to IndexFlatL2 but uses an inverted file structure for faster search.
2. **faiss.IndexIVFPQ**: This is an implementation of Product Quantization (PQ) that can handle large-scale datasets with low memory usage. It performs coarse quantization using an inverted file structure and then applies PQ encoding on each subvector.
3. **faiss.IndexIVFADC**: This is an implementation of Additive Quantization (AQ) that can handle large-scale datasets with low memory usage. It performs coarse quantization using an inverted file structure and then applies AQ encoding on each subvector.
4. **faiss.IndexHNSWFlat**: This is an implementation of Hierarchical Navigable Small World (HNSW) algorithm that uses a graph-based approach for efficient nearest neighbor search.
5. **faiss.IndexIVFSpectralHash**: This is an implementation of Spectral Hashing that uses the eigenvalues of a graph Laplacian matrix to quantize the data points. It can handle large-scale datasets with low memory usage.
6. **faiss.IndexLSH**: This is an implementation of Locality Sensitive Hashing that uses hash functions to map similar data points to the same bucket for faster search. It can handle large-scale datasets with low memory usage

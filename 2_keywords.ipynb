## YAKE

https://github.com/LIAAD/yake<br>
https://liaad.github.io/yake/docs/-getting-started<br>

The lower the score, the more relevant the keyword is.

```python
import yake

text = """La scientologie est un ensemble de croyances et de pratiques dont les principes ont été développés aux États-Unis en 1952 par L. Ron Hubbard. L'Église de Scientologie fut fondée à Camden dans le New Jersey et ouvrit en décembre 1953. La Scientologie promeut une méthode pseudoscientifique appelée dianétique par son fondateur et propose plus largement un ensemble de croyances et de pratiques relatives à la nature de l'être humain et de sa place dans l'Univers, enseignant que les humains sont des êtres immortels qui ont oublié leur véritable nature.
Elle est considérée comme une secte ou comme une religion selon le pays. De plus, certains pays la considèrent comme une entreprise commerciale, d'autres comme étant une organisation à but non lucratif ou charitable. Sa classification fait parfois l'objet de contentieux.
Plusieurs organisations encadrant la pratique de la Scientologie ont été créées, la plus notable étant l'Église de Scientologie. L'Église de Scientologie internationale (en), dont le siège est à Los Angeles aux États-Unis, est considérée comme étant l'Église mère des organisations et des entités de la Scientologie."""

language = "fr"
max_ngram_size = 3
deduplication_threshold = 0.9
numOfKeywords = 100
custom_kw_extractor = yake.KeywordExtractor(lan=language, n=max_ngram_size, dedupLim=deduplication_threshold, top=numOfKeywords, features=None)
custom_kw_extractor_sw = yake.KeywordExtractor(lan=language, n=max_ngram_size, dedupLim=deduplication_threshold, top=numOfKeywords, features=None, stopwords="CSV/stopwords-fr.csv")

keywords = custom_kw_extractor.extract_keywords(text)
keywords_sw = custom_kw_extractor_sw.extract_keywords(text)

print("\tAvec stopwords\t\t\t\t\t\t Sans stopwords")
print("----------------------------------------------------------------------------------------------")
for i in range(20):
    kw, score = keywords[i]
    kw_sw, score_sw = keywords_sw[i]
    print(f"{kw:30} ({score:.3f})\t\t\t{kw_sw:30} ({score_sw:.3f})")
```

## PKE

https://github.com/boudinfl/pke

```python
import pke

# initialize keyphrase extraction model, here TopicRank
extractor = pke.unsupervised.TopicRank()

# load the content of the document, here document is expected to be a simple 
# test string and preprocessing is carried out using spacy
extractor.load_document(input='text', language='en')

# keyphrase candidate selection, in the case of TopicRank: sequences of nouns
# and adjectives (i.e. `(Noun|Adj)*`)
extractor.candidate_selection()

# candidate weighting, in the case of TopicRank: using a random walk algorithm
extractor.candidate_weighting()

# N-best selection, keyphrases contains the 10 highest scored candidates as
# (keyphrase, score) tuples
keyphrases = extractor.get_n_best(n=10)
```

**KeyWords without and without lemmatization**

```python
def extract_keywords(txt, numOfKeywords, max_ngram_size):
    custom_kw_extractor = yake.KeywordExtractor(lan=language, n=max_ngram_size, dedupLim=deduplication_threshold, top=numOfKeywords, features=None)
    keywords = custom_kw_extractor.extract_keywords(txt)
    return [kw[0] for kw in keywords]

df_keywords = df_lemm.copy()
df_keywords['kw1'] = df_keywords['Documents'].apply(lambda txt : extract_keywords(txt, 10, 1))
# df_keywords['kw2'] = df_keywords['Documents'].apply(lambda txt : extract_keywords(txt, 10, 2))
df_keywords['kw3'] = df_keywords['Documents'].apply(lambda txt : extract_keywords(txt, 10, 3))
df_keywords['kw1_Lemm'] = df_keywords['Documents_Lemm'].apply(lambda txt : extract_keywords(txt, 10, 1))
# df_keywords['kw2_lemm'] = df_keywords['Documents_Lemm'].apply(lambda txt : extract_keywords(txt, 10, 2))
df_keywords['kw3_Lemm'] = df_keywords['Documents_Lemm'].apply(lambda txt : extract_keywords(txt, 10, 3))
df_keywords.to_csv('CSV/documents_keywords.csv')
# df_keywords[['kw1', 'kw2', 'kw3']]
df_keywords[['kw1', 'kw3', 'kw1_Lemm', 'kw3_Lemm']]
# df_keywords
```
